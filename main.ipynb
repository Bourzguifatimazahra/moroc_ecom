```python
import pandas as pd
import numpy as np
 
# Charger le dataset

df = pd.read_csv("morocco_ecommerce.csv",sep="|")

# Nettoyer les noms de colonnes : enlever espaces et mettre en minuscules
df.columns = df.columns.str.strip().str.lower()
print("Colonnes du dataset :\n", df.columns.tolist())
 
# Dataset Structure & Basic Exploration

print("10 premières lignes :\n", df.head(10))
print("Forme du dataset (rows, columns) :\n", df.shape)
print("Types de données :\n", df.dtypes)
 
# Missing Values and  Data Quality

missing_counts = df.isna().sum() #fct isna().sum() refais 
print("Valeurs manquantes par colonne :\n", missing_counts)
max_missing = missing_counts.max()
print("Colonnes avec le plus de NaN :\n", missing_counts[missing_counts == max_missing])


# Remplir les NaN dans city et unit_price
if 'city' in df.columns:
    df['city'] = df['city'].fillna("Unknown")
if 'unit_price' in df.columns:
    df['unit_price'] = df['unit_price'].fillna(df['unit_price'].mean())

# Supprimer toutes les lignes avec NaN
df_no_na = df.dropna()
print("Nombre de lignes après suppression des NaN :\n", df_no_na.shape[0])

 
# Duplicates

print("Lignes dupliquées complètes :\n", df.duplicated().sum())
df = df.drop_duplicates()
print("Nombre de lignes après suppression des doublons :", df.shape[0])

if 'order_id' in df.columns:
    df = df.drop_duplicates(subset='order_id', keep='first')

# Filtering & Subsetting

if 'quantity' in df.columns:
    print("Commandes quantity > 3 :\n", df[df['quantity'] > 3])
if 'total_amount' in df.columns:
    print("Commandes total_amount > 1000 :\n", df[df['total_amount'] > 1000])
if 'region' in df.columns:
    print("Commandes Casablanca-Settat :\n", df[df['region'] == "Casablanca-Settat"])
if 'payment_method' in df.columns:
    print("Commandes non payées en Cash :\n", df[df['payment_method'] != "Cash"])

#  Descriptive Statistics

if 'total_amount' in df.columns:
    print("Stats total_amount :\n", df['total_amount'].agg(['mean', 'median', 'min', 'max']))

if 'region' in df.columns and 'total_amount' in df.columns:
    region_avg = df.groupby('region')['total_amount'].mean()
    print("Région avec le plus haut total_amount moyen :\n", region_avg.sort_values(ascending=False).head(1))

if 'product_id' in df.columns and 'total_amount' in df.columns:
    product_revenue = df.groupby('product_id')['total_amount'].sum()
    print("Produit avec le plus haut total revenue :\n", product_revenue.sort_values(ascending=False).head(1))

# String Cleaning

if 'city' in df.columns:
    df['city'] = df['city'].str.strip().str.title()
    df['city'] = df['city'].replace({'Casa': 'Casablanca', 'Cas': 'Casablanca'})

# Date & Time Features

if 'order_date' in df.columns:
    df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
if 'ship_date' in df.columns:
    df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce')

if 'order_date' in df.columns:
    df['year'] = df['order_date'].dt.year
    df['month'] = df['order_date'].dt.month
    df['weekday'] = df['order_date'].dt.day_name()

 # Commandes après le 2023-06-01
    orders_after_june = df[df['order_date'] > '2023-06-01']
    print("Commandes après 2023-06-01 :\n", orders_after_june)
 
 # Moyenne mensuelle du total_amount
    if 'total_amount' in df.columns:
        df.set_index('order_date', inplace=True)
        monthly_revenue = df['total_amount'].resample('ME').mean()  # 'ME' = month end, future-proof
        print("Moyenne mensuelle du total_amount :\n", monthly_revenue)

#  Outlier Detection

if 'total_amount' in df.columns:
    # Méthode IQR
    Q1 = df['total_amount'].quantile(0.25)
    Q3 = df['total_amount'].quantile(0.75)
    IQR = Q3 - Q1
    outliers_iqr = df[(df['total_amount'] < Q1 - 1.5*IQR) | (df['total_amount'] > Q3 + 1.5*IQR)]
    print("Outliers détectés par IQR :\n", outliers_iqr)

    # Méthode Z-score (sans SciPy)
    z_scores = (df['total_amount'] - df['total_amount'].mean()) / df['total_amount'].std()
    outliers_z = df[abs(z_scores) > 3]
    print("Outliers détectés par Z-score :\n", outliers_z)

    # 5 plus grandes commandes
    top5_orders = df.nlargest(5, 'total_amount')[['order_id', 'total_amount']]
    print("5 plus grandes commandes :\n", top5_orders)
```
